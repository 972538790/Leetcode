[162. Find Peak Element](https://leetcode.com/problems/find-peak-element/)
```java
class Solution {
    public int findPeakElement(int[] nums) {
        // 此处套用二分法的模板即可
        int L = 0;
        int R = nums.length - 1; // 此处之所以要减1，是因为在guess函数中会+1，防止其越界。
        int ans = 0;
        while(L < R) {
            int mid = (L + R) / 2;
            if(guess(mid, nums)) {
                // 通常ans=mid，此处由于mid的下一位才可能为正确答案（nums[mid] < nums[mid+1]），所以ans=mid+1
                ans = mid + 1;
                L = mid + 1;
            }
            else {
                R = mid;
            }
        }
        return ans;
    }
    public boolean guess(int mid, int[] nums) {
        // 算法的精髓 当nums[mid+1]>nums[mid]时说明右边一定存在正确答案，因为nums[n] = -∞
        return nums[mid] < nums[mid+1];
    }
}
```

[72. Edit Distance](https://leetcode.com/problems/edit-distance/)
```java

class Solution {
    public static int[][] temp;
    public static int min(int b, int c, int d) {
        return Math.min(Math.min(b, c), d);
    }
    public int solve(String word1, String word2, int idx1, int idx2) {
        if(idx1 < 0 || idx2 < 0) {
            return idx1 < 0? idx2+1: idx1+1;
        }

        if(idx1 < 0 && idx2 < 0) {
            return 0;
        }

        if(temp[idx1][idx2] != -1) return temp[idx1][idx2];
        if(word1.charAt(idx1) == word2.charAt(idx2)) {
            temp[idx1][idx2] = solve(word1, word2, idx1-1, idx2-1);
            
        }
        else {
            // 添加一个数
            int b = solve(word1, word2, idx1, idx2-1) + 1;
            // 替换一个数
            int c = solve(word1, word2, idx1 -1, idx2 - 1) + 1;
            // 删除一个数
            int d = solve(word1, word2, idx1 - 1, idx2) + 1;
            int res = min(b, c, d);
            temp[idx1][idx2] = res;
        }
        return temp[idx1][idx2];
    }
    public int minDistance(String word1, String word2) {
        temp = new int[word1.length()][word2.length()];
		// for (int[] item: temp)
		// {
		// 	Arrays.fill(item, -1); // 为每一维数组赋初始值
		// }
        for(int i = 0; i < temp.length; i++) {
            for(int j = 0; j < temp[i].length; j++) {
                temp[i][j] = -1;
            }
        }
        return solve(word1, word2, word1.length()-1, word2.length()-1);
    }
}
```

```java
class Solution {
    public int[] twoSum(int[] nums, int target) {
        HashMap<Integer, Integer> hashMap = new HashMap<>();
        for(int i = 0; i < nums.length; i++) {
            hashMap.put(nums[i], i);
        }
        
        for(int i = 0; i < nums.length; i++) {
            int left = target - nums[i];
            if(hashMap.containsKey(left) && hashMap.get(left) != i) {
                return new int[] {i, hashMap.get(left)};
            }
        }
        
        throw new IllegalArgumentException("No two sum solution");
    }
}
```

```java
            add = 1;
        }

        l2 = l2.next;

            current.next = new ListNode(add);
        }
        
        while(l1 != null && l2 != null) {
            
            
            if(l1.val + l2.val + add >= 10) {
                current.next = new ListNode((l1.val + l2.val + add)%10);
                // 进位
                add = 1;
            }
            else {
                current.next = new ListNode(l1.val + l2.val + add);
                add = 0;
            }
            l1 = l1.next;
            l2 = l2.next;
            current = current.next;
        }
        while(l1 != null) {
            current.next = new ListNode((l1.val + add) % 10);
            if(l1.val + add >= 10) {
                add = 1;
            }
            else {
                add = 0;
            }
            // 
            
            l1 = l1.next;
            current = current.next;
        }
        while(l2 != null) {
            current.next = new ListNode((l2.val + add) % 10);
            if(l2.val + add >= 10) {
                add = 1;
            }
            else {
                add = 0;
            }
            l2 = l2.next;
            current = current.next;

            current.next = new ListNode(add);
```

```java
public class Solution {
    public int lengthOfLongestSubstring(String s) {
        int n = s.length();
        int start = 0;
        int end = 0;
        int ans = 0;
        HashSet<Character> set = new HashSet<Character>();
        while(start < n && end < n) {
            if(!set.contains(s.charAt(end))) {
                set.add(s.charAt(end++));
                ans = Math.max(ans, end - start);
            }
            else {
                set.remove(s.charAt(start++));
            }
        }
        return ans;
    }
```

```c++
        if (!root) return false;
        if (!root->left && !root->right)   return pasum + root->val == sum;
        return dfs(root->left, pasum + root->val, sum) || dfs(root->right, pasum + root->val, sum);
    }

        return dfs(root, 0, sum);
    }
```

[892. Surface Area of 3D Shapes](https://leetcode.com/problems/surface-area-of-3d-shapes/)

````python
class Solution(object):
    def surfaceArea(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        res = 0
        for i in range(len(grid)):
            for j in range(len(grid[i])):
                if grid[i][j] == 0:
                    continue
                area = 2 + 4 * grid[i][j]
                if i-1 >= 0:
                    area -= min(grid[i-1][j],grid[i][j])
                if j-1 >= 0:
                    area -= min(grid[i][j],grid[i][j-1])
                if i + 1 < len(grid):
                    area -= min(grid[i][j],grid[i+1][j])
                if j + 1 < len(grid[i]):
                    area -= min(grid[i][j],grid[i][j+1])
                res += area
        return res 
````

```c++
class Solution {
public:
    bool isPrime(int num) {
        if (num <= 1) return false;
        for (int i = 2; i*i <= num; ++i) {
            if (num%i == 0) return false;
        }
        return true;
    }

    int countPrimes(int n) {
        bool *isPrime = new bool[n];
        for (int i = 2; i < n; ++i) {
            isPrime[i] = true;
        }
        for (int i = 2; i*i < n; ++i) {
            if (!isPrime[i]) continue;
            for (int j = i*i; j < n; j += i) {
                isPrime[j] = false;
            }
        }
        int count = 0;
        for (int i = 2; i < n; ++i) {
            if (isPrime[i]) count++;
        }
        return count;
    }
};
```

```java
class Solution {
    public int uniquePathsWithObstacles(int[][] obstacleGrid) {
        int m = obstacleGrid.length;
        int n = obstacleGrid[0].length;
        int[][] memory = new int[m][n];
        for(int i = 0; i < m; i++) {
            for(int j = 0; j < n; j++) {
                if(obstacleGrid[i][j] == 1) {
                    if(i == 0 && j == 0) {
                        return 0;
                    }
                    memory[i][j] = 0;
                    continue;
                }
                if(i==0 && j==0) {
                    memory[i][j] = 1;
                }
                else if(i == 0) {
                    memory[i][j] = memory[i][j-1];
                }
                else if(j == 0) {
                    memory[i][j] = memory[i-1][j];
                }
                else {
                    memory[i][j] = memory[i-1][j] + memory[i][j-1];
                }
            }
        }
        return memory[m-1][n-1];
    }
}
```

```java
            }
        }

    }
```

```java
class Solution {
    public boolean guess(int mid, int[] nums) {
        int count = 0;
        for(int i = 0; i < nums.length; i++) {
            if(mid >= nums[i]) count++;
        }
        return count <= mid;
    }
    public int findDuplicate(int[] nums) {
        int n = nums.length - 1;
        int min = 1;
        int max = n;
        int ans = 0;
        while(min <= max) {
            int mid = min + (max- min) / 2;
            if(guess(mid, nums)) {
                min = mid + 1;               
            }
            else {
                ans = mid;
                max = mid- 1;
            }
        }
        return ans;
        
    }
}
```
